#3.Medias móviles simples

**Suavizamiento de una serie**: Acercar datos al centro, eliminar impacto de valores extremos. Así es más fácil ver una tendencia porque se elimina el ruido blanco de la serie.


**Características de MA:** Las medias moviles se usan como suavizadores, sirven para estimar la tendencia de una serie cuando no es posible describirla en ecuaciones parámetricas simples como regresión lineal. Este tipo de métodos para estimar la tendencia también se conocen como **filtros lineales**, la estimación de la tendencia se haría por medio del calculo de la estimación de media móvil

Las medias moviles no solo se usan para estimar la tendencia, también se utiliza para hacer pronóstico. En este caso, es más recomendable para series estacionarias.



**Medias móviles bilaterales:** Se usan cuando se van a utilizar como filotros lineales para extraer el componente de tendencia de la serie, el calculo de las medias móviles es el siguiente: 

<br>
<br>

**$$\hat{y_{t}}=\frac{(y_{t+n}+..y_{t+2}+y_{t+1}+y_{t}+y_{t-1}+y_{t-2}+...+ y_{t-n})}{2n+1}$$**

<br>

El parámetro n se conoce como orden, ancho de ventana o ventana.

ej:

Media móvil para el periodo 5 con una ventana de 2 periodos.

**$$\hat{y_{5}}=\frac{(y_{7}+y_{6}+y_{5}+y_{4}+y_{3})}{5}$$**



**Media móvil unilaterales** Se usan cuando se quiere pronósticar basado en medias móviles esta técnica consiste en calcular el valor de una serie con base en el promedio de n periodos anteriores.

<br>
<br>

**$$\hat{y_{t}}=\frac{(y_{t-1}+y_{t-2}+...+ y_{t-n})}{n}$$**



```r
install.packages("smooth")  ### paquete para usar técnicas de suavizamiento para pronóstico sma
#install.packages("forecast") ### función de medias moviles ma
library(smooth)
#library(forecast)
```

```r
set.seed(100)
t=c(1:100)
y=ts(3+5*t+rnorm(100,200,50),start=1500) ## se crea una serie con tendencia


ma_y=sma(y,h=25,silent = F,interval=T, level=0.90) ## se ajusta un modelo de medias móviles

class(ma_y)
summary(ma_y) 
```

**residuals:** Resta entre valores ajustados y valores reales

# 4.Suavizamiento exponencial simple:

Es un tipo de filtro lineal similar a medias móviles, pero que utiliza una ponderación exponencial. Es decir los factores de ponderación decaen exponencialmente. Esto se utiliza para darle más peso a los valores de la serie recientes y menos peso a los valores de la serie más antiguos.

El pronóstico basado en suavizamiento exponencial simple se recomienda para series estacionarias, sin embargo hay variaciones del suavizamiento exponencial (modelos Holt y modelos Holt-Winters) para incluir los componentes de tendencia y estacionalidad. (se verán más adelante)

La ecuación que representa este modelo es la siguiente.


**$$\hat{y_{t}}=\alpha *y_{t-1}+(1-\alpha )*\hat{y_{t-1}}$$**


**$$\hat{y_{t}}=\alpha *y_{t-1}+(1-\alpha )(\alpha*y_{t-2} +(1-\alpha )*\hat{y_{t-2}})$$**




**$$\hat{y_{t}}=\alpha *y_{t-1}+(1-\alpha )\alpha*y_{t-2} +(1-\alpha )^{2}*\hat{y_{t-2}}$$**

## 5.4 Teoría de suavizamiento exponential de Holt

En 1957 Charles Holt propone una extensión del modelo de suavizamiento exponencial para que permita el pronóstico de series con tendencia.

El modelo de pronóstico incluye dos ecuaciones de suavizamiento, una para la tendencia y otra ecuación para el nivel.

pronóstico del periodo t+h:
 $$   \hat{y}_{t+h|t}= \ell_{t}+hb_{t}$$

 Ecuación de nivel ($\ell_{t}$):

 $$\ell_{t} = \alpha y_{t}+(1-\alpha)(\ell_{t-1}+ b_{t-1}) $$

 Ecuacuión de tendencial ($ b_{t}$):

 $$ b_{t}=\beta(\ell_{t}-\ell_{t-1})+(1-\beta)b_{t-1}$$


$\alpha$ y $\beta$ son los parametros de suavizamiento de la ecuación de nivel y de la ecuación de tendencia respectivamente toman valores entre 0 y 1 (Funcionan igual que el parámetro del suavizamiento exponencial).

La ecuación de nivel es similar a suavizamiento exponencial, que utiliza un promedio ponderado entre el valor actual de la serie y la predicción del periodo anterior.

La ecuación de tendencia es un promedio ponderado entre la tendencia estimado en el periodo actual y la tendencia estimadaen en el periodo anterior.

##6.1Modelos de holt-winters Teoría

Este modelo fue propuesto en dos trabajos presentados por charles Holt (1957) y Peter Winter (1960). En estos trabajos se buscaba hacer una extensión del modelo de suavizamiento exponencial de holt para que se pudieran modelar series con estacionalidad. 

Este modelo tiene una ecuación de pronóstico y tres ecuaciones de suavizamiento, para el nivel, para la tendencia y para la estacionalidad.

Existen dos variantes del componente estacional.

1. Aditivo cuando la variación en la estacionalidad (picos) es constante a lo largo de la serie.
2. Multiplicativo cuando la variación de la estacionalidad en la serie va aumentando a medida que aumenta el nivel (valor promedio) de la serie.


**Aditivo**

Ecuación para el pronóstico:

\begin{align*}
  \hat{y}_{t+h|t} &= \ell_{t} + hb_{t} + s_{t+h-m(k+1)} \\\\\end{align*}

  Ecuación de nivel:
\begin{align*}
 \ell_{t} &= \alpha(y_{t} - s_{t-m}) + (1 - \alpha)(\ell_{t-1} + b_{t-1})
 \\\\\end{align*}

Ecuación de tendencia
\begin{align*}
  b_{t} &= \beta^*(\ell_{t} - \ell_{t-1}) + (1 - \beta^*)b_{t-1}
  \\\\\end{align*}

Ecuación de estacionalidad
\begin{align*}
  s_{t} &= \gamma (y_{t}-\ell_{t-1}-b_{t-1}) + (1-\gamma)s_{t-m},
\end{align*}


Los componentes nuevos son: $\gamma$ que es el parámetro de suavizamiento de la estacionalidad, m que es el número de periodos en el que se repite el ciclo, $k= (h-1/m)$ que permite asegurar que la estimación de la estacionalidad viene de la última observación de los datos reales.


**Multiplicativo**


\begin{align*}
  \hat{y}_{t+h|t} &= (\ell_{t} + hb_{t})s_{t+h-m(k+1)} \\
  \ell_{t} &= \alpha \frac{y_{t}}{s_{t-m}} + (1 - \alpha)(\ell_{t-1} + b_{t-1})\\
  b_{t} &= \beta^*(\ell_{t}-\ell_{t-1}) + (1 - \beta^*)b_{t-1}                \\
  s_{t} &= \gamma \frac{y_{t}}{(\ell_{t-1} + b_{t-1})} + (1 - \gamma)s_{t-m}
\end{align*}

```r
### Mape de entrenamiento 1 periodo
ajustado=sma_model$fitted
real=austourists

mape(real,ajustado)
```

```r
#### Para mape de evaluación h periodos de pronóstico
library(smooth)
sma_f<-function(x,h){  ### se debe crear función con la función de pronóstico y los argumentos que se quieran utilizar
  
  greybox::forecast(sma(x,h,order=8))
}


e=tsCV(austourists,sma_f,h=10)  ### esta función permite hacer el calculo de varios pronósticos de un mismo h al mismo tiempo

e

erroresh1<-e[29:48,1]
erroresh2<-e[29:48,2]
erroresh3<-e[29:48,3]
erroresh10<-e[29:48,10]

mapeh1=mean(abs(erroresh1/ausair[29:48]),na.rm=T)
mapeh2=mean(abs(erroresh2/ausair[29:48]),na.rm=T)
mapeh3=mean(abs(erroresh3/ausair[29:48]),na.rm=T)
mapeh10=mean(abs(erroresh10/ausair[29:48]),na.rm=T)

mapeh1
mapeh2
mapeh3
mapeh10
```

```r
## MAPE Entrenamiento 1 periodo
ajustado=ses_model$fitted
real=ses_model$x

mape(real,ajustado) ## debe coincidir con salidad del modelo
```
